```python
df["embarked"] = df["embarked"].astype("category")
str(df["embarked"].dtype)
df.info()
```

```python
df[df["embarked"]=="C"].head(10)
```

```python
df[df["embarked"]!="S"].head()
```

```python
df[(df["fare"]>500)|(df["age"]>70)].head()
```

```python
df.isnull().sum()
```

```python
df = df.drop("who", axis=1)
df.columns
```

```python
df["deck"]=df["deck"].fillna(df["deck"].mode()[0])
df["deck"]
df.head()
```

```python
df["deck"].isnull().values.any()
```

```python
df["age"].isnull().values.any()
```

```python
df["age"]=df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
df.groupby(["pclass","sex"]).agg({'survived':['sum','count','mean']})
```

```python
df["age_flag"]=df["age"].apply(lambda x:1 if x<30 else 0)
df["age_flag"].head()
```

```python
df = sns.load_dataset("tips")
df.head()
df.shape
```

```python
df.groupby("time").agg({"total_bill":["sum","min","max","mean"]})
```

```python
df.groupby(["day","time"]).agg({"total_bill":["sum","min","max","mean"]})
```

```python
df.loc[(df["time"] == "Lunch") & (df["sex"]=="Female")]
df.groupby(["day"]).agg({"total_bill" :["sum","min","max"],"tip":["sum","min","max"]})
```

```python
df.loc[(df["size"] < 3) & (df["total_bill"] >10 ) , "total_bill"].mean()
```

```python
df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
df.head()
```

```python
new_df2 = df.sort_values("total_bill_tip_sum", ascending=False)[:30]
new_df2.head(20)
```
